```python
import pandas as pd

df1 = pd.read_parquet('../data/DENGBR17.parquet')
df2 = pd.read_parquet('../data/DENGBR18.parquet')
df3 = pd.read_parquet('../data/DENGBR19.parquet')

df = pd.concat([df1, df2, df3])
```

```python
minhas_colunas = [
    'DT_PCR', 'UF', 'DT_OBITO', 'RESUL_PRNT', 'MANI_HEMOR', 'ARTRITE',
    'HEMATOLOG', 'RES_CHIKS1', 'COUFINF', 'RESUL_SORO', 'MIALGIA',
    'DT_PRNT', 'GRAV_ORGAO', 'DT_NS1', 'DT_INVEST', 'ALRM_ABDOM',
    'SG_UF_NOT', 'IMUNOH_N', 'SOROTIPO', 'GRAV_INSUF', 'ACIDO_PEPT',
    'CONJUNTVIT', 'EPISTAXE', 'FLXRECEBI', 'DOR_COSTAS', 'TP_NOT',
    'ID_MUNICIP', 'SEM_NOT', 'DT_INTERNA', 'HIPERTENSA', 'HEPATOPAT',
    'ID_MN_RESI', 'ALRM_LIQ', 'DT_SIN_PRI', 'METRO', 'TP_SISTEMA',
    'HISTOPA_N', 'CS_ESCOL_N', 'GRAV_CONSC'
]

df = df[minhas_colunas]
```

```python
# Removendo colunas que assumem dengue
col = [
    "SOROTIPO",
    "COUFINF",
    "DT_NS1",
    "DT_PCR",
    "DT_OBITO",
    "DT_PRNT",
    "RESUL_PRNT",
    "RES_CHIKS1",
    "RESUL_SORO",
    "ALRM_ABDOM",
    "ALRM_LIQ",
    "GRAV_ORGAO",
    "GRAV_CONSC",
    "GRAV_INSUF",
    "HISTOPA_N",
    "IMUNOH_N"
]

df = df.drop(columns=col)
```

```python
# Removes columns that have all null values or only one unique value
def remove_same_val(df):
    for column in df.columns:
        if df[column].isnull().all():
            print(f"Column '{column}' has all null values.")
            df = df.drop(column, axis=1)
        elif df[column].nunique() == 1:
            print(f"Column '{column}' has only one unique value: {df[column].unique()[0]}")
            df = df.drop(column, axis=1)
    return df

df = remove_same_val(df)
```

```python
# Doencas preexistentes
col = [
    "HEMATOLOG",
    "HEPATOPAT",
    "HIPERTENSA",
    "ACIDO_PEPT"
]

df.drop(columns=col)
```

```python
# Colunas a serem descartadas

col = [
    "DT_INVEST",
    "DT_INTERNA",
    "SG_UF_NOT",
    "TP_NOT", # 99.9 da mesma classe
]

df = df.drop(columns=col)
```
